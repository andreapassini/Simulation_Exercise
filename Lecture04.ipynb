```python
horses = list( range(6) )
horses
```

```python
import itertools
```

```python
S = list( itertools.permutations(horses) )
```

```python
S
```

```python
# event A: horse 3 preceeds horse 4 at the finish
A = list( filter( lambda L: L.index(3) < L.index(4),   S)  ) 
```

```python
A
```

```python
len(S)
```

```python
len(A)
```

```python
# B: horse 3 arrives in position 1
# C: horse 3 arrives in position 2
B = list( filter(lambda L: L.index(3) == 1, S) )
C = list( filter(lambda L: L.index(3) == 2, S) )
```

```python
B
```

```python
C
```

```python
# D: AcapB: event A and B
AcapB = list ( set.intersection(set(A), set(B)) )
```

```python
AcapB
```

```python
# E: event A or B (either horse 3 arrives before 4 OR 3 is in position 1 - or both)
AcupB = list ( set.union( set(A), set(B) ) )

# events D and E are intertwined

# F: event B and C (horse 3 arrives in position 1 AND in position 2)
# is empty: B and C are *mutually exclusive*
BcapC = list ( set.intersection(set(B), set(C)) )
# G: event B or C (horse 3 arrives in position 1 OR in position 2)
# finding its probability is easy: P[B] + P[C]
BcupC = list ( set.union( set(B), set(C) ) )
```

```python
print( len(S), len(A), len(B), len(C), len(AcapB), len(AcupB), len(BcapC), len(BcupC) )
```

```python
# Example of a (discrete) RANDOM VARIABLE: X -> arrival position of horse 3
# Possible Events for X: 3 arrives in position 0, 3 arrives in position 1, ...
# Example of another (discrete) RANDOM VARIABLE: Y -> arrival position of horse 5
# Possible Events for Y: 5 arrives in position 0, 5 arrives in position 1, ...
# X and Y are *independent*: whichever events C and D I choose for X and Y, resp., 
# P[X \in C, Y \in D] = P[X \in C] * P[Y \in D]
```

```python
# Example of a (discrete) RANDOM VARIABLE: X' -> horse 3 wins
# Possible Events for X': 3 wins (arrives in position 0), 3 does not win
# Example of another (discrete) RANDOM VARIABLE: Y' -> horse 5 wins
# Possible Events for Y': 5 wins (arrives in position 0), 5 does not win
# X' and Y' are *NOT independent*: if I take an event C for X' and and event D for Y'
# P[X' \in C, Y' \in D] != P[X' \in C] * P[Y' \in D]
# e.g. P[X' \in {3 wins}] = 1/6; P[Y' \in {5 wins}] = 1/6; P[X' \in {3 wins}, Y' \in {5 wins}] = 0
```
